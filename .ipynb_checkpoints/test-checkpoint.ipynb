```python
file = 'logs/Graph Classification/Eval_Other_Worms_10/output.pk'
with open(file, 'rb') as pickle_file:
    content = pk.load(pickle_file)
```

```python
import numpy as np
from itertools import combinations
keys = [list(combinations([1,2,3,4,5], i+1)) for i in range(5)]
GNN_results = {}
MLP_results = {}
for key, value in content.items():
    split_name = key.split(' | ')
    assert split_name[0] in ['GNN_Att', 'MLP']
    if split_name[0]=='GNN_Att':
        GNN_results.update({split_name[1]:value})
    elif split_name[0]=='MLP':
        MLP_results.update({split_name[1]:value})
```

```python
GNN_train_1 = []
for element in keys[0]:
    GNN_train_1.extend(GNN_results[str(element)]['Kato_corr'])
buf = np.array(GNN_train_1)
print('1:', np.mean(buf), np.std(buf))
GNN_train_2 = []
for element in keys[1]:
    GNN_train_2.extend(GNN_results[str(element)]['Kato_corr'])
buf = np.array(GNN_train_2)
print('2:', np.mean(buf), np.std(buf))
GNN_train_3 = []
for element in keys[2]:
    GNN_train_3.extend(GNN_results[str(element)]['Kato_corr'])
buf = np.array(GNN_train_3)
print('3:', np.mean(buf), np.std(buf))
GNN_train_4 = []
for element in keys[3]:
    GNN_train_4.extend(GNN_results[str(element)]['Kato_corr'])
buf = np.array(GNN_train_4)
print('4:', np.mean(buf), np.std(buf))
```
