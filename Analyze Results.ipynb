```python
import numpy as np
from itertools import combinations
keys = [list(combinations([1,2,3,4,5], i+1)) for i in range(5)]
GNN_traj = {}
MLP_traj = {}
for key, value in content_traj.items():
    split_name = key.split(' | ')
    assert split_name[0] in ['GNN_Att', 'MLP','Node']
    if split_name[0] in ['GNN_Att','Node']:
        GNN_traj.update({split_name[1]:value})
    elif split_name[0]=='MLP':
        MLP_traj.update({split_name[1]:value})
```

```python
def traj_mse(arg):
    buf = []
    buf_std = []
    pred_steps = 16
    for element in arg:
        size = element.size(1)
        if (size%pred_steps) != 0:
            end = (size%pred_steps)
            buf_1=np.mean(np.array(element.cpu()),axis=(0,-1))[:-end].reshape(-1,pred_steps)
        else:
            buf_1 =np.mean(np.array(element.cpu()),axis=(0,-1)).reshape(-1,pred_steps)
        buf.append(np.mean(buf_1,axis=0))
        buf_std.append(np.std(buf_1,axis=0))
    return np.mean(np.stack(buf,-1),axis=-1), np.std(np.stack(buf,-1),axis=-1)
```
