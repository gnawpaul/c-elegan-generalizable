```python
def calculate_matrix(results):
    pred_1 = []
    tar_1 = []
    for element in keys[0]:
        pred_1.extend(results[str(element)]['results']['predicted_states'])
        tar_1.extend(results[str(element)]['results']['ground_truth'])
    mat=np.zeros((5,5))
    for pred, tar in zip(pred_1, tar_1):
        mat[int(pred)][int(tar)] +=1
    return mat[:4,:4]
```

```python
def combine_turns(matrix):
    cor = 0
    for i in range(4):
        for j in range(4):
            if i==j:
                cor+=matrix[i,j]
            elif i==3 and j ==2:
                cor += matrix[i,j]
            elif j==3 and i ==2:
                cor += mat[i,j]
    return cor/np.sum(matrix, axis=(0,1))
```

```python
def traj_mse(arg):
    buf = []
    buf_std = []
    pred_steps = 16
    batches = 0
    for element in arg:
        size = element.size(1)
        if (size%pred_steps) != 0:
            end = (size%pred_steps)
            buf_1=np.mean(np.array(element.cpu()),axis=(0,-1))[:-end].reshape(-1,pred_steps)
        else:
            buf_1 =np.mean(np.array(element.cpu()),axis=(0,-1)).reshape(-1,pred_steps)
        batches += buf_1.shape[0]
        buf.append(np.mean(buf_1,axis=0))
        buf_std.append(np.std(buf_1,axis=0))
    print(batches)
    return np.mean(np.stack(buf,-1),axis=-1), np.std(np.stack(buf,-1),axis=-1)
```
