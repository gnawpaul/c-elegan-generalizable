```python
import numpy as np
from itertools import combinations
keys = [list(combinations([1,2,3,4,5], i+1)) for i in range(5)]
GNN_results = {}
MLP_results = {}
for key, value in content.items():
    split_name = key.split(' | ')
    assert split_name[0] in ['GNN_Att', 'MLP']
    if split_name[0]=='GNN_Att':
        GNN_results.update({split_name[1]:value})
    elif split_name[0]=='MLP':
        MLP_results.update({split_name[1]:value})
```
